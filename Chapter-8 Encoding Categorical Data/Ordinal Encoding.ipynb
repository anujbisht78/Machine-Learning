```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('customer.csv')
```

```python
df.sample(50)
```

```python
df=df.iloc[:,2:]
```

```python
df.head(50)
```

```python
#whenever we do feature transformation, first we do Train_Test_Split 
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
x_train
```

Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
#creating an object for Ordinal Encoder
#then in the object we ll create a parameter which in a list and inside the list, we will pass the list of categories we want to convert.
#the list will depend upon the number of columns
#if i dont provide the order then kisi ko bhi random number mil jata
```

```python
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
#now to train the data set
oe.fit(x_train)
```

```python
X_train=oe.transform(x_train)
X_test=oe.transform(x_test)
```

```python
X_train
```

```python
#converting into datarame
X_train=pd.DataFrame(X_train,columns=x_train.columns)
X_test=pd.DataFrame(X_test,columns=x_test.columns)
```
