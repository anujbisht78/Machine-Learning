```python
import pandas as pd
import numpy as np
```

Pandas Series

```python
l=[1,2,3,4,5]
newl=pd.Series(l)
newl
```

```python
#iterating the items
print(newl[0])
print(newl[3])
print(newl[4])
```

```python
#creating own Indexing
li=[1,2,3,4,5]
nl=pd.Series(li,index=['x','y','z','a','d'])
nl
```

```python
#now iterate
print(nl['x'])
print(nl['z'])
print(nl['d'])
```

Data Frame

```python
#Series are like Columns and DataFrame are like whole Table
col={"Name":["Ajay","Rahal","Mohan"],"Age":[25,45,85],"Place":["Delhi","UK","Himanchal"]}
newcol=pd.DataFrame(col)
newcol
```

```python
#Indicating particular Row
newcol.loc[1]
```

```python
#if we want to extract any coloumn or any row
# Extraction one coloumn
newcol.iloc[:,1:2]
```

```python
newcol.iloc[:,2:3]
```

```python
#Extractionj one row
newcol.iloc[1:2,:]
```
