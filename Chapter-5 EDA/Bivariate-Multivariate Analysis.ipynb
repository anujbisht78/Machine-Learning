```python
#when we want to analise 2 columns together, we have to check the data types of both the columns
#there are three ways:
# 1. Numerical-Numerical
# 2. Numerical-Categorical
# 3. Categorical-Categorical
```

1. Scatterplot(Numerical-Numerical)

```python
sns.scatterplot(tips['total_bill'],tips['tip'])
```

```python
#there is a linear sort of relationship
```

```python
#now we will do multivariate Analysis
#in this scatterplot i have to find the sex ratio
```

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['smoker'])
#it is the multivariate Analysis where we are talking about 3 quantities
```

2. Bar Plot(Numerical-Categorical)

```python
#we generally use barplot when one data is numerical and one data is categorical
# x-axis-categorical and y-axis-numerical
sns.barplot(titanic['Pclass'],titanic['Age'])
```

```python
#now performing multivariate Analysis
sns.barplot(titanic['Pclass'],titanic['Age'],hue=titanic['Sex'])
```

3. BoxPlot (Numerical-Categorical)

```python
sns.boxplot(titanic['Sex'],titanic['Age'])
```

```python
sns.boxplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
```

4. Distplot (Numerical-categorical)

```python
#performing the pdf
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

5. HeatMap(Categorical-Categorical)

```python
titanic.head()
```

```python
#i want to find ho many people died and survived in each Pclass
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

```python
#ploting the series
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
titanic.groupby('Sex').mean()['Survived']*100
```

6. clusterMap(Categorical-Categorical)

```python
#cluster map tells the closeness between the values
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```
